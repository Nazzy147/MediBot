```python
# ============================================================================
# SECTION 3: SEVERITY CHECKER CLASS
# ============================================================================

print("\n" + "="*80)
print("Creating Severity Assessment System...")
print("="*80)

class SeverityChecker:
    """Assess symptom severity and provide urgency levels"""

    def __init__(self):
        self.severe_keywords = {
            'headache': ['worst headache', 'sudden severe', 'thunderclap',
                        'with fever and stiff neck', 'vision loss', 'confusion',
                        'worst of my life'],
            'fever': ['above 104', '105', 'seizure', 'difficulty breathing',
                     'chest pain', 'severe headache with fever', 'stiff neck',
                     'unusual rash', 'over 40 celsius'],
            'stomach': ['severe pain', 'bloody stool', 'black stool',
                       'vomiting blood', 'can\'t keep fluids', 'severe cramping',
                       'unbearable pain'],
            'chest': ['chest pain', 'chest pressure', 'pain radiating',
                     'shortness of breath', 'cold sweat', 'arm pain with chest'],
            'breathing': ['can\'t breathe', 'cannot breathe', 'difficulty breathing',
                         'gasping', 'choking']
        }

        self.moderate_keywords = {
            'headache': ['frequent', 'several days', 'getting worse', 'daily',
                        'three days', 'week'],
            'fever': ['3 days', '4 days', 'five days', 'not responding',
                     'high fever'],
            'stomach': ['persistent', 'several days', 'unable to eat',
                       'constant pain']
        }

    def assess_severity(self, user_message):
        """Assess severity level"""
        message_lower = user_message.lower()

        # Emergency check
        emergency_phrases = [
            'chest pain', 'can\'t breathe', 'cannot breathe',
            'difficulty breathing', 'severe bleeding', 'unconscious',
            'seizure', 'stroke', 'suicide', 'overdose', 'choking',
            'heart attack', 'anaphylaxis'
        ]

        for phrase in emergency_phrases:
            if phrase in message_lower:
                return 'emergency', 'ğŸš¨ EMERGENCY: Call 108 immediately!'

        # Severe check
        for category, keywords in self.severe_keywords.items():
            for keyword in keywords:
                if keyword in message_lower:
                    return 'urgent', 'ğŸŸ  URGENT: See a doctor today or visit urgent care.'

        # Moderate check
        for category, keywords in self.moderate_keywords.items():
            for keyword in keywords:
                if keyword in message_lower:
                    return 'moderate', 'ğŸŸ¡ MODERATE: Schedule doctor appointment within 2-3 days.'

        return 'mild', 'ğŸŸ¢ MILD: Try home remedies. See doctor if symptoms persist/worsen.'

    def get_severity_badge(self, severity_level):
        """Return colored badge"""
        badges = {
            'emergency': 'ğŸ”´ EMERGENCY',
            'urgent': 'ğŸŸ  URGENT',
            'moderate': 'ğŸŸ¡ MODERATE',
            'mild': 'ğŸŸ¢ MILD'
        }
        return badges.get(severity_level, 'ğŸŸ¢ MILD')

print("âœ“ Severity checker created!")
```

```python
# ============================================================================
# SECTION 4: DURATION TRACKER CLASS
# ============================================================================

print("\n" + "="*80)
print("Creating Symptom Duration Tracker...")
print("="*80)

class SymptomTracker:
    """Track symptom duration and patterns"""

    def __init__(self):
        self.symptoms_log = {}

    def extract_duration(self, message):
        """Extract duration from message"""
        message_lower = message.lower()

        # Numeric patterns
        patterns = [
            (r'(\d+)\s*days?', 'days'),
            (r'(\d+)\s*weeks?', 'weeks'),
            (r'(\d+)\s*hours?', 'hours'),
            (r'(\d+)\s*months?', 'months')
        ]

        for pattern, unit in patterns:
            match = re.search(pattern, message_lower)
            if match:
                return {'value': int(match.group(1)), 'unit': unit}

        # Relative terms
        if 'yesterday' in message_lower or 'since yesterday' in message_lower:
            return {'value': 1, 'unit': 'days'}
        elif 'few days' in message_lower or 'couple days' in message_lower:
            return {'value': 3, 'unit': 'days'}
        elif 'week' in message_lower:
            return {'value': 1, 'unit': 'weeks'}

        return None

    def assess_chronicity(self, duration_info):
        """Assess if acute or chronic"""
        if not duration_info:
            return 'acute', "Duration unclear - treating as new symptom."

        value = duration_info['value']
        unit = duration_info['unit']

        # Convert to days
        if unit == 'weeks':
            days = value * 7
        elif unit == 'months':
            days = value * 30
        elif unit == 'hours':
            days = value / 24
        else:
            days = value

        if days < 3:
            return 'acute', f"Recent symptom ({value} {unit})."
        elif days < 14:
            return 'subacute', f"Ongoing for {value} {unit} - consider seeing doctor if not improving."
        else:
            return 'chronic', f"Chronic symptom ({value} {unit}) - please consult healthcare provider."

    def generate_duration_advice(self, message):
        """Generate advice based on duration"""
        duration_info = self.extract_duration(message)
        chronicity, message_text = self.assess_chronicity(duration_info)

        advice = {
            'acute': "\n\nâ° **Duration**: Recent symptom - monitor for 2-3 days.",
            'subacute': "\n\nâ° **Duration**: Ongoing symptom - schedule doctor visit soon.",
            'chronic': "\n\nâ° **Duration**: Long-standing - requires medical evaluation."
        }

        return f"\nğŸ“Š **Timeline**: {message_text}{advice.get(chronicity, '')}"

print("âœ“ Duration tracker created!")
```

```python
# ============================================================================
# SECTION 5: MULTILINGUAL SUPPORT
# ============================================================================

print("\n" + "="*80)
print("Creating Multilingual Support...")
print("="*80)

class MultilingualSupport:
    """Multi-language translation support"""

    def __init__(self):
        try:
            self.translator = Translator()
            self.supported_languages = {
                'en': 'English', 'hi': 'Hindi', 'es': 'Spanish',
                'fr': 'French', 'de': 'German', 'pt': 'Portuguese',
                'bn': 'Bengali', 'ta': 'Tamil', 'te': 'Telugu', 'mr': 'Marathi'
            }
            self.enabled = True
        except:
            self.enabled = False
            print("âš ï¸  Translation service unavailable - using English only")

    def detect_language(self, text):
        """Detect language"""
        if not self.enabled:
            return 'en'
        try:
            detection = self.translator.detect(text)
            return detection.lang
        except:
            return 'en'

    def translate_to_english(self, text, source_lang):
        """Translate to English"""
        if not self.enabled or source_lang == 'en':
            return text
        try:
            translation = self.translator.translate(text, src=source_lang, dest='en')
            return translation.text
        except:
            return text

    def translate_from_english(self, text, target_lang):
        """Translate from English"""
        if not self.enabled or target_lang == 'en':
            return text
        try:
            translation = self.translator.translate(text, src='en', dest=target_lang)
            return translation.text
        except:
            return text

multilingual = MultilingualSupport()
print("âœ“ Multilingual support initialized!")
```

```python
# ============================================================================
# SECTION 7: DATA PREPROCESSING
# ============================================================================

print("\n" + "="*80)
print("Processing Medical Data...")
print("="*80)

class MedicalChatbotPreprocessor:
    """Data preprocessing"""

    def __init__(self):
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))
        self.stop_words.discard('not')
        self.stop_words.discard('no')
        self.stop_words.discard('very')
        self.words = []
        self.classes = []
        self.documents = []
        self.ignore_words = ['?', '.', ',', '!', "'", '"', '(', ')', '[', ']']

    def preprocess_text(self, text):
        """Preprocess text"""
        tokens = word_tokenize(text.lower())
        tokens = [self.lemmatizer.lemmatize(word) for word in tokens
                 if word.isalnum() and word not in self.ignore_words]
        return tokens

    def load_and_process_data(self, intents_file='medical_intents.json'):
        """Load and process data"""
        with open(intents_file, 'r') as f:
            intents = json.load(f)

        for intent in intents['intents']:
            for pattern in intent['patterns']:
                word_list = self.preprocess_text(pattern)
                self.words.extend(word_list)
                self.documents.append((word_list, intent['tag']))
                if intent['tag'] not in self.classes:
                    self.classes.append(intent['tag'])

        self.words = sorted(list(set(self.words)))
        self.classes = sorted(list(set(self.classes)))

        print(f"âœ“ Processed {len(self.documents)} patterns")
        print(f"âœ“ Vocabulary: {len(self.words)} words")
        print(f"âœ“ Classes: {len(self.classes)} categories")

        return intents

    def create_training_data(self):
        """Create training data"""
        training = []
        output_empty = [0] * len(self.classes)

        for doc in self.documents:
            bag = []
            pattern_words = doc[0]

            for word in self.words:
                bag.append(1 if word in pattern_words else 0)

            output_row = list(output_empty)
            output_row[self.classes.index(doc[1])] = 1
            training.append([bag, output_row])

        random.shuffle(training)
        training = np.array(training, dtype=object)

        train_x = np.array(list(training[:, 0]))
        train_y = np.array(list(training[:, 1]))

        return train_x, train_y

preprocessor = MedicalChatbotPreprocessor()
intents_data = preprocessor.load_and_process_data()
train_x, train_y = preprocessor.create_training_data()
```

```python
# ============================================================================
# SECTION 8: BUILD AND TRAIN MODEL
# ============================================================================

print("\n" + "="*80)
print("Building Neural Network...")
print("="*80)

np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

def create_model(input_shape, num_classes):
    """Create neural network"""
    model = keras.Sequential([
        layers.Dense(256, input_shape=(input_shape,), activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

model = create_model(len(train_x[0]), len(train_y[0]))
print("âœ“ Model created!")

print("\nTraining model...")
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_accuracy',
    patience=20,
    restore_best_weights=True,
    verbose=0
)

lr_scheduler = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=0.00001,
    verbose=0
)

history = model.fit(
    train_x, train_y,
    epochs=300,
    batch_size=8,
    validation_split=0.15,
    verbose=0,
    callbacks=[early_stopping, lr_scheduler],
    shuffle=True
)

print(f"âœ“ Training complete!")
print(f"  Final accuracy: {history.history['accuracy'][-1]:.4f}")
print(f"  Best validation: {max(history.history['val_accuracy']):.4f}")

# Save model and data
model.save('medimitra_model.h5')
pickle.dump(preprocessor.words, open('words.pkl', 'wb'))
pickle.dump(preprocessor.classes, open('classes.pkl', 'wb'))
pickle.dump(preprocessor, open('preprocessor.pkl', 'wb'))
print("âœ“ Model and data saved!")
```

```python
# ============================================================================
# SECTION 9: ENHANCED CHATBOT CLASS
# ============================================================================

print("\n" + "="*80)
print("Creating Enhanced Chatbot System...")
print("="*80)

class EnhancedMedicalChatbot:
    """Complete enhanced medical chatbot"""

    def __init__(self, model_path='medimitra_model.h5'):
        self.model = keras.models.load_model(model_path)
        self.words = pickle.load(open('words.pkl', 'rb'))
        self.classes = pickle.load(open('classes.pkl', 'rb'))
        self.preprocessor = pickle.load(open('preprocessor.pkl', 'rb'))
        self.intents = intents_data
        self.severity_checker = SeverityChecker()
        self.symptom_tracker = SymptomTracker()
        self.multilingual = multilingual
        self.interactive_checker = interactive_checker

        print("âœ“ MediBot  initialized!")
        print(f"  Knowledge base: {len(self.classes)} categories")
        print(f"  Vocabulary: {len(self.words)} words")

    def clean_up_sentence(self, sentence):
        """Clean sentence"""
        return self.preprocessor.preprocess_text(sentence)

    def bag_of_words(self, sentence):
        """Create bag of words"""
        sentence_words = self.clean_up_sentence(sentence)
        bag = [0] * len(self.words)
        for w in sentence_words:
            for i, word in enumerate(self.words):
                if word == w:
                    bag[i] = 1
        return np.array(bag)

    def predict_class(self, sentence):
        """Predict intent"""
        bow = self.bag_of_words(sentence)
        res = self.model.predict(np.array([bow]), verbose=0)[0]
        max_index = np.argmax(res)
        return [{
            "intent": self.classes[max_index],
            "probability": str(res[max_index])
        }]

    def get_response(self, intents_list):
        """Get response"""
        if not intents_list:
            return "I couldn't understand that. Could you rephrase your question?"

        tag = intents_list[0]["intent"]
        confidence = float(intents_list[0]["probability"])

        for intent in self.intents["intents"]:
            if intent["tag"] == tag:
                response = random.choice(intent["responses"])
                if confidence < 0.6:
                    response += "\n\n(Note: I'm not entirely certain. Please consult a healthcare professional for accurate advice.)"
                return response

        return "I can help with medical questions. What's your health concern?"

    def chat(self, message, user_id="default_user", language='auto'):
        """Main chat function with all enhancements"""

        if not message or not message.strip():
            return "Please ask me a medical question!"

        # Language detection/translation
        detected_lang = 'en'
        if language == 'auto' and self.multilingual.enabled:
            detected_lang = self.multilingual.detect_language(message)
            if detected_lang != 'en':
                message = self.multilingual.translate_to_english(message, detected_lang)

        # Emergency check
        emergency_keywords = ['emergency', 'urgent', 'chest pain', 'cannot breathe',
                            'can\'t breathe', 'severe bleeding', 'unconscious',
                            'overdose', 'heart attack', 'stroke', 'choking']

        if any(k in message.lower() for k in emergency_keywords):
            response = "ğŸš¨ MEDICAL EMERGENCY! ğŸš¨\n\nIf this is a medical emergency, call 108 (Emergency Services) IMMEDIATELY or go to the nearest ER!\n\nâš¡ Time is critical - don't delay! This chatbot cannot provide emergency care! ğŸ¥"
            if detected_lang != 'en':
                response = self.multilingual.translate_from_english(response, detected_lang)
            return response

        # Unsupported check
        unsupported_keywords = ['rash', 'skin', 'acne', 'eye', 'vision', 'ear',
                               'broken', 'fracture', 'arthritis', 'pregnancy',
                               'cancer', 'tumor', 'asthma', 'kidney', 'liver']

        if any(k in message.lower() for k in unsupported_keywords):
            response = ("This condition is outside my scope. I specialize in: headaches, fever, "
                       "cold/flu, stomach issues, diabetes, blood pressure, mental health basics, "
                       "and general wellness.\n\nFor your concern, please consult a qualified "
                       "healthcare professional.")
            if detected_lang != 'en':
                response = self.multilingual.translate_from_english(response, detected_lang)
            return response

        # Predict intent and get response
        intents = self.predict_class(message)
        base_response = self.get_response(intents)

        # Add severity assessment
        severity_level, severity_rec = self.severity_checker.assess_severity(message)
        severity_badge = self.severity_checker.get_severity_badge(severity_level)

        # Add duration analysis
        duration_advice = self.symptom_tracker.generate_duration_advice(message)

        # Combine responses
        enhanced_response = f"{severity_badge}\n{severity_rec}\n\n{'-'*60}\n\n{base_response}{duration_advice}"

        # Add disclaimer
        disclaimer = "\n\nâš ï¸ **Disclaimer:** This is general health information only. Always consult qualified healthcare professionals for medical advice, diagnosis, or treatment."

        final_response = enhanced_response + disclaimer

        # Translate back if needed
        if detected_lang != 'en' and self.multilingual.enabled:
            final_response = self.multilingual.translate_from_english(final_response, detected_lang)

        return final_response

chatbot = EnhancedMedicalChatbot()
print("âœ“ MediBot  ready!")
```
